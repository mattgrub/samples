# **Section 4: Data Science**

#### Null hypothesis: The variables influencing wine quality are consistent across red and white wines.

#### Alternative hypothesis: Specific factors (such as volatile acidity, sulfates, or others) have a stronger impact on the quality of red wines compared to white wines, and vice versa.

#### This hypothesis comes from the differences in the most correlated variables between red and white wines. 

### First, I am going to add a variable to each dataset to determine if it is red or white wine.

```python
redwinelabeled = redwine.copy()  
redwinelabeled['class'] = 0
redwinelabeled
```

```python
whitewinelabeled = whitewine.copy()  
whitewinelabeled['class'] = 1
whitewinelabeled
```

All wines with class equaling 0 are red, and all wines with class equaling 1 are white.

### Creating three datasets

```python
twofiftyred = redwinelabeled.nlargest(250, "quality")
twofiftywhite = whitewinelabeled.nlargest(250, "quality")
```

The above datasets contain the 250 highest quality wines for red and white respectively.

```python
evenbest = pd.concat([twofiftyred, twofiftywhite], ignore_index=True)
```

The above dataset, **evenbest**, contains the top 250 best red and white wines combined into a single dataset. This will be used later for visualizations.

### I now will create correlation matrices for the 250 best red and white wines separately. 

Correlation Matrix of Best Red Wines

```python
red_wine_corr = twofiftyred.corr()
red_wine_corr
```

Top Three Most Correlated Variables with Quality:
1. Alcohol: .4703
2. Citric Acid: .2474
3. Chlorides: -.2294

Correlation Matrix of Best White Wines

```python
white_wine_corr = twofiftywhite.corr()
white_wine_corr
```

Top Three Most Correlated Variables with Quality: 
1. Alcohol: .1948
2. Residual Sugars: .1443 
3. Volatile Acidity: .1258

#### It is interesting to see that other than alcohol, all of the most correlated variables in the top 250 datasets are different from the redwine and whitewine datasets. Also, it is shown that white wine consistently has lower levels of correlation across all variables. 

## Correlation Heat Maps for each type of wine

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
sns.heatmap(red_wine_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix - Red Wine')

plt.subplot(1, 2, 2)
sns.heatmap(white_wine_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix - White Wine')

plt.tight_layout()
plt.show()
```

# **Section 5: Overall Summary**

#### Overall, this analysis has supported the alternative hypothesis that specific factors support quality in each type of wine, with the exception of alcohol content. Across all parts of analysis, alcohol has been the most correlated factor to a high quality wine, with increased alcohol percentage leading to better scores. Outside of alcohol, there was no consistency in factors correlating with quality. Both red and white wine had completely different top 3 correlating factors, and also the datasets containing the 250 best wines had completely different top 3 correlating factors from their respective full datasets. This shows that outside of alcohol there are no factors that can be used as predictors of a quality wine.

#### Outside of our hypothesis, there were some interesting findings we can take away from this analysis. The overall correlation between quality and all other factors is quite low for white wine. This can mean two things to me, either the scoring for quality of white wine had a wider variance and comes down more to subject biases, or that it is possible to make white wine taste good in many different ways. Red wine on the otherhand had more correlation between quality and certain factors such as acidity and sulphates, relative to white wine. This could be because of any number of factors, and could prove to be a great point for further analysis.
